4.1 Задайте массив случайных значений из интервала (0; 1). Рассчитайте средние и медианные значения для массива, сравните результаты, какие выводы можно сделать о значениях? Постройте точечную диаграмму рассения полученного ряда.

```python
from random import randint
import matplotlib.pyplot as plt
from statistics import median
data = [i / 10 for i in range(10)]
massiv = []
for i in range(1000):
      massiv.append(data[randint(0,9)])
av = sum(massiv)/len(massiv)
med = median(massiv)
for i in range(10):
      data[i] = massiv.count(i/10)
print(av, med)
if (av > med):
      print("Среднее значение больше медианного")
elif (av == med):
      print("Среднее значение равно медианному")
else:
      print("Среднее значение меньше медианного")
plt.scatter(sorted(set(massiv)), data)
plt.show()
```

4.2 Дана функция
image.png Создать массив из 10 значений функции (𝑥, например, изменяется от 1 до 10). Выделить срез первой половины массива и построить графики для основного массива – линейный и для среза – точечный.


```python
import math as m
import matplotlib.pyplot as plt
data, marks = [], []
for x in range(1, 11):
    data.append( (m.sqrt(1 + (m.e)**(m.sqrt(x)) + m.cos(x**2)) / abs(1 - (m.sin(x))**3)) + ((m.log(abs(2*x))) / (m.log(m.e))) )
    marks.append(x)
data_half = data[:5]
marks_half = marks[:5]
plt.plot(marks, data)
plt.scatter(marks_half, data_half)
plt.show()
```

4.3 Дана функция image.png Построить график на интервале (0,10) с шагом 1 с заливкой площади и найти эту площадь под ним. Для вычисления площади воспользуйте функции trapz(), вычисляющей интеграл по правилу трапеции. Для ее корректной работы необходимо подключить следующие
библиотеки:

> from scipy.integrate import simps

> from numpy import trapz

```python
import numpy as np
import matplotlib.pyplot as plt
data, marks = [], []
for x in range(10):
    data.append(np.abs(np.cos(x * np.e ** (np.cos(x) + (np.log(x+1) / np.log(np.e))))))
    marks.append(x)
plt.grid()
plt.plot(marks, data, c = "r")
plt.fill_between(marks, data)
plt.show()
print(np.trapz(data))
```

4.4 Откройте статистические данные по стоимости акций Apple, Microsoft, Google за 12 месяцев 2021 г. Постройте 3 графика на плоскости и оцените их динамику. Подумайте, как можно улучшить читаемость результатов. Для начальных данных допустимо использовать следующий ресурс: https://finance.yahoo.com

```python
import matplotlib.pyplot as plt
file = list(open('AIS_1.csv'))
for i in range(13):
    file[i] = file[i].replace(',', ' ')
    file[i] = file[i].replace('\n', '')
    file[i] = list(file[i].split(" "))
marks, Apple, Microsoft, Google = [], [], [], []
for i in range(1, 13):
    for j in range(4):
        file[i][j] = float(file[i][j])
        if j == 0:
            marks.append(file[i][j])
        if j == 1:
            Apple.append(file[i][j])
        if j == 2:
            Microsoft.append(file[i][j])
        if j == 3:
            Google.append(file[i][j])

plt.plot(marks, Apple, c = "r")
plt.plot(marks, Microsoft, c = "g")
plt.plot(marks, Google, c = "b")
plt.legend(['Apple', 'Microsoft', 'Google'])
plt.show()
```
