1.1 Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и единицами.

```python
import numpy as np
chess = np.zeros((8,8),dtype=int)
chess[1::2,::2] = 1
chess[0::2,1::2] = 1
print(chess)
```

1.2 Создать 5x5 матрицу со значениями в строках от 0 до 4. Для создания необходимо использовать функцию arrange.


```python
import numpy as np
matrix = np.zeros((5,5),int)
matrix += np.arange(5,dtype=int)
print(matrix)
```

1.3 Создать массив 3x3x3 со случайными значениями.

```python
import numpy as np
matrix = np.random.random((3,3,3))*10
for i in range(3):
    for j in range(3):
        for k in range(3):
            matrix[i][j][k] = int(matrix[i][j][k])
print(matrix)
```

1.4 Создать матрицу с 0 внутри, и 1 на границах.

```python
import numpy as np
matrix = np.ones((10,10),dtype=int)
matrix[1:-1, 1:-1] = 0
print(matrix)
```

1.5 Создайте массив и отсортируйте его по убыванию.

```python
import numpy as np
length = 25
matrix = np.random.random((length))
matrix = sorted(matrix, reverse=True)
for i in range(length):
    matrix[i] = int(100*matrix[i])
print(matrix)
```

1.6 Создайте матрицу, выведите ее форму, размер и размерность.

```python
import numpy as np
a, b = int(input('Кол-во строк: ')), int(input('Кол-во столбцов: '))
matrix = np.random.random((a,b))
for i in range(a):
    for j in range(b):
        matrix[i][j] = int(100*matrix[i][j])
print(matrix)
print(np.shape(matrix))
print(np.size(matrix))
print(np.ndim(matrix))
```

2.1 Найдите евклидово расстояние между двумя Series (точками) a и b, не используя встроенную формулу.

3.2 Загрузить фрейм данных по ссылке: https://raw.githubusercontent.com/akmand/datasets/master/iris.csv. Необходимо выполнить нормализацию первого числового признака (sepal_length_cm) с использованием минимаксного преобразования, а второго (sepal_width_cm) с задействованием z-масштабирования.

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
url = 'https://raw.githubusercontent.com/akmand/datasets/master/iris.csv'
dataframe = pd.read_csv(url)
scaler = MinMaxScaler()
dataframe[['sepal_length_cm']] = scaler.fit_transform(dataframe[['sepal_length_cm']])
scaler = StandardScaler()
dataframe[['sepal_width_cm']] = scaler.fit_transform(dataframe[['sepal_width_cm']])
dataframe
```
